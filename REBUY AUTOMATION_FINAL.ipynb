## Calculate T_total

```python
# 2. T_total

# #SKU is unique
# #ASIN amazon serial id is not necessarily unique 


# if asin appears multiple times, change the total in stock to reflect the total actually in stock
# for example if asin appears 2x and row 1 has total in stock = 1 and row 2 has total in stock = 25, update the row of t_toal to 26 for both row 1 and row 2
# if it shows up 2x (same asin but 2 different rows), that means it comes from 2 separate vendors (different sku) and they have different prices, so for another sheet can decide if you want to buy again
# but here, you just want to know what the total inventory is... 
# because for later, you want to know how much from vendor 1 and how much from vendor 1, the prices that each vendor sold it for, so you can decide who to buy it from and how much is in stock 
# this should be a different cleaned up dataframe, in theory 

df['total instock'] = df['Current Inventory'] + df['Prepping Units for Amazon'] + df['Inbound At Shipper Units'] + df['Inbound At Amazon Receiving Units'] + df['Transfer Units']
df['TTotal'] = df.groupby(['ASIN'])['total instock'].transform('sum')

x = df.ASIN.unique()
x.shape

dup = df[df.duplicated(subset=['ASIN'])]

dup = dup.sort_values(by='ASIN')
```

## Calculate Days in Stock

```python
df['days instock'] = (df['In Stock at Amazon Last 1 - 30 Day(s)']*0.01)*30
df['days instock'] = df['days instock'].replace([0],6)
```

## Calculate Sellable Quantity SLBL_QTY
## DEFAULT TO 6 if the result of SLBL QTY is 1-5.99 just = to 6

```python
df.loc[(df['FBA Sales 30 Days'] >= 15) & (df['Rank'] <= 800), 'SLBL QTY'] = ((df['FBA Sales 30 Days']/df['days instock'])*55) - df['TTotal']
df.loc[(df['FBA Sales 30 Days'] >= 15) & (df['Rank'] >= 801), 'SLBL QTY'] = ((df['FBA Sales 30 Days']/df['days instock'])*30) - df['TTotal']

df.loc[(df['FBA Sales 30 Days'] < 15), 'SLBL QTY'] = (df['FBA Sales 30 Days']*2) - df['TTotal']
df.loc[(df['FBA Sales 30 Days'] == 0) | pd.isnull(df['FBA Sales 30 Days']), 'SLBL QTY'] = (df['FBA Sales 90 Days']/2)- df['TTotal']
df.loc[(df['SLBL QTY'] >= 1) & (df['SLBL QTY'] <= 5.99), 'SLBL QTY'] = 6
```

## Convert DataTypes to Float and Remove Money sign

```python
df['New, 3rd Party FBA: 90 days avg.'] = df['New, 3rd Party FBA: 90 days avg.'].replace({'\$':''}, regex = True)
df['New, 3rd Party FBA: 90 days avg.'] = df['New, 3rd Party FBA: 90 days avg.'].replace({',':''}, regex = True)

# Removing Dollar sign and converting to float
convert_to_float = ['New, 3rd Party FBA: 90 days avg.', 'lowest_new_price_fba']
for c in convert_to_float:
    df[c] = df[c].replace('$ ', '')
    #df[c] = df[c].astype(float)
    
convert_to_float = ['New, 3rd Party FBA: 90 days avg.', 'lowest_new_price_fba']
new_var_name = ['new_third_party_fba', 'new_low_price_fba']
for c,n in [(convert_to_float, new_var_name) for c in convert_to_float for n in new_var_name]:
    
    df[n] = df[c].replace({'\$':''}, regex = True)
    df[n] = df[n].replace({',':''}, regex = True)

    df[n] = df[n].astype(float)
    df[n] = df[n].replace(0, np.nan)   
    
    
# FBA Equal Checks

def fba_equal_check(row):
    
    if row['new_third_party_fba'] ==  row['new_low_price_fba']:
        return 'equal'
    elif row['new_third_party_fba'] < row['new_low_price_fba']:
        return 'third'
    elif row['new_third_party_fba'] > row['new_low_price_fba']:
        return 'low'
    else:
        return ''
        
df['fba_check'] = df.apply(fba_equal_check, axis=1)
```

## Calculate BuyBox Prices

```python
def buy_box_price(row):
    if row['new_third_party_fba'] > row['new_low_price_fba'] and row['buybox_new_landed_price'] == 0 or pd.isnull(row['buybox_new_landed_price']) :
        return row['new_low_price_fba']
    elif row['new_third_party_fba'] < row['new_low_price_fba'] and row['buybox_new_landed_price'] == 0 or pd.isnull(row['buybox_new_landed_price']) :
        return row['new_third_party_fba']
    elif  row['buybox_new_landed_price'] == 0 and pd.notnull(row['new_third_party_fba']) and  pd.isnull(row['new_low_price_fba']):
        return row['new_third_party_fba']
    
    elif  row['buybox_new_landed_price'] == 0 and pd.isnull(row['new_third_party_fba']) and  pd.notnull(row['new_low_price_fba']):
        return row['new_low_price_fba']
    elif pd.notnull(row['new_third_party_fba'] ) and pd.isnull(row['new_low_price_fba']) and pd.isnull(row['buybox_new_landed_price']):
        return row['new_third_party_fba']
    

    
    elif pd.isnull(row['new_third_party_fba'] ) and pd.notnull(row['new_low_price_fba']) and pd.isnull(row['buybox_new_landed_price']):
        return row['new_low_price_fba']

    
    
    
    
    
    elif pd.notnull(row['buybox_new_landed_price']):
        #return row['buybox_new_landed_price']
        return row['buybox_new_landed_price']
   

    
df['BB_PRICE'] = df.apply(buy_box_price, axis=1)
```

## Flag Empty Buy Box Price

```python
def flag_price_empty(row):
    if pd.isnull(row['new_third_party_fba']) and pd.isnull(row['new_low_price_fba']) and pd.isnull(row['buybox_new_landed_price']):
        return 'no price values'
    elif pd.isnull(row['new_third_party_fba']) and pd.isnull(row['new_low_price_fba']) and row['buybox_new_landed_price'] == 0:
        return 'price is zero'
    elif row['new_third_party_fba'] == 0 or row['new_low_price_fba'] == 0 or row['buybox_new_landed_price'] == 0:
        return 'price is zero'
    
    
    
df['flag_BB_PRICE'] = df.apply(flag_price_empty, axis=1)
```

## Calculate referral fee for specific categories

```python
def f(row):
    
    if row['BB_PRICE'] <= 10 and row['sales_rank_category'] == 'baby product' or row['sales_rank_category'] == 'beauty' or row['sales_rank_category'] == 'health and beauty':
        return 0.08
    elif row['sales_rank_category'] == 'biss':
        return 0.12
    elif row['sales_rank_category'] == 'fashion' or row['sales_rank_category'] == 'watches':
        return 0.16
    elif row['sales_rank_category'] == 'grocery' and row['BB_PRICE'] <=15:
        return 0.08
    else:
       return 0.15
        
df['referral_fee'] = df.apply(f,axis=1)
```

## Calculates List Pack from SKU

```python
# LP = list pack (if pack of 6, LP =6, if single, LP = 1)

# LP comes from the sku, it comes from the last 3 digits - 100
df['skulen'] = df['SKU'].astype(str).map(len)
print(df['skulen'].value_counts())
df['SKU'].loc[(df.skulen == 21)]

#ignore dash, just take sku 19 and 20 <-- this not true - ask michael 

#get last item in string of SKU, put it in a list, and then just get the unique values 
sku_endings = list(set(df['SKU'].str[-1:].tolist()))


# convert numbers to float while leaving strings in string format b/c want a list of just strings 

def maybe_float(s):
    try:
        return float(s)
    except (ValueError, TypeError):
        return s


sku_endings = [maybe_float(v) for v in sku_endings]


#get just the string endings 
sku_endings_str = [ x for x in sku_endings if type(x)==str ]


def f(row):
    if row['skulen'] == 19:
        return row['SKU'][16:]
    elif row['skulen'] == 20 and row['SKU'].endswith(tuple(sku_endings_str)):
        return row['SKU'][16:19]
    
    elif row['skulen'] == 20 :
        return row['SKU'][17:20]
    
df['LP'] = df.apply(f,axis=1)
```
